```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train, epochs=120)
```

```python
model.metrics_names
```

```python
Loss = model.history.history['loss']
```

```python
sns.lineplot(x=range(len(Loss)), y=Loss)
```

```python
prediction = model.predict(X_test)
```

```python
training_score = model.evaluate(X_train, y_train, verbose=0)
```

```python
test_score = model.evaluate(X_test, y_test, verbose=0)
```

```python
print(f"training_score: {training_score}\ntest_score: {test_score}")
```

```python
training_score - test_score
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print(f"mean_squared_error: {mean_squared_error(y_test, prediction)}")
print(f"root_mean_squared_error: {np.sqrt(mean_squared_error(y_test, prediction))}")
print(f"mean_absolute_error: {mean_absolute_error(y_test, prediction)}")
```
